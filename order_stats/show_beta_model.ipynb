```python
import numpy as np
import pandas as pd
from scipy.stats import beta
from matplotlib import pyplot as plt
```

```python
points = np.linspace(5.7,23,100)
```

```python
a = np.float(2.5)
b = np.float(4)
loc = np.float(5.7)
scale = np.float(19)

plt.plot(points, beta.cdf(points, a, b, loc, scale))
plt.xlabel("per mile pace")
plt.ylabel("cumulative probability")
plt.title("cumulative probability for per mile pace from beta model")
plt.savefig("cumulative probability for per mile pace from beta model", transparent=True)
```
