```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
m_win = np.array([0.800, 0.768, 0.911, 0.922, 0.955])
g_win = np.array([0.200, 0.072, 0.089, 0.047, 0.045])
tie   = np.array([0.000, 0.160, 0.000, 0.031, 0.000])
pop   = np.array([1    , 2    , 3    , 4    , 5])
```

```python
fig, axs = plt.subplots(1, 1)
fig.suptitle("probability of each outcome with different number of runners")

axs.plot(pop, m_win, label="M win prob", c="orange")
axs.plot(pop, g_win, label="G win prob", c="steelblue")
axs.plot(pop, tie, label="tie prob", c="green")

plt.xticks(pop)
plt.xlabel("number of runners from each school")
plt.ylabel("probability")
plt.legend()

plt.savefig('prob-pop.png', transparent=True)

fig.show()
```
