# Generating the empirical distribution
This notebook was written by Yubo Wang. It contains jupyter notebook code for generating the empirical distribution in the XC Running.
## General Question:
Other than the ideal distribution, we need to generated an empirical distribution that based on the beta model of per-mile pace for runners that introduced in the previous research. 
## Our Approach:
We simulate the per mile pace for each runner from each school and rank them from top 1 to top 16 (this part is done in other ntoebook). Then we use the frequency of each school being into the top 16 devided by the total time of simulation to get teh probability of each school being into the top 16.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### DataFrames:
#### frequency of each school being into the top 16 (167$\times$3)
result
* ID (int): unique school ID number
* pop (int): number of girl students in school
* freq (int): frequency of being into the top 16 in the simulation

```python
result = pd.read_csv(r"top16_freq.csv")
result["prob"] = result["freq"] / result["freq"].sum() * 16
```

```python
plt.scatter(result["pop"], result["prob"], label="mean", s=3)
plt.xlabel("population")
plt.ylabel("probability")
plt.title("probability of being into the top 16 for each school")
plt.savefig('prob_vs_pop2', transparent=True)
plt.show()
```

```python
result = result[["ID", "pop", "prob"]]
result.to_csv("empirical_distribution16.csv", index=False)
```
